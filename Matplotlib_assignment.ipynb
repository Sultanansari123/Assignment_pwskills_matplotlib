# Ans1-

Matplotlib is a python library for data visualization.It is used to create static,animated and interactive plots

 Five types of plots include line plots, scatter plots, bar plots, histogram plots, and pie charts.






# Ans2-

A scatter plot is a type of plot that displays individual data points on a two-dimensional graph

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for demonstration

x = np.random.rand(100)
y = 2 * x + 1 + 0.1 * np.random.randn(100)

# Plot the scatter plot
plt.scatter(x, y, label='Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')
plt.legend()
plt.show()
```

# Ans3-

The subplot() function in Matplotlib is used to create multiple plots in the same figure. Each subplot represents a separate plot within the same figure. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for four lines
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.subplot(2,2, 1)
plt.plot(x, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line Plot 4')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

# Ans4-

comparision between categorical and numerical data then used bar plots.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit, color='blue')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.title('Bar Plot: Company Profit')
plt.show()

# Horizontal bar plot
plt.barh(company, profit, color='green')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.title('Horizontal Bar Plot: Company Profit')
plt.show()
```

# Ans5-

A box plot, also known as a box-and-whisker plot, is a graphical summary of a set of data values.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data into a list
data = [box1, box2]

# Create a box plot
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot: Comparison of Box 1 and Box 2')
plt.show()
```

Box plots are useful for detecting outliers and comparing distributions.

It displays the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum
